```python
# 71

import pandas as pd

df_links = pd.read_csv('../data/links.csv', index_col=0)
df_links.head()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline

G = nx.Graph()

for column in df_links.columns:
    node_no = column.strip('Node')
    G.add_node(str(node_no))

for idx, df_link in df_links.iterrows():
    node_no = idx.strip('Node')
    for inner_idx, item in df_link.iteritems():
        _node_no = inner_idx.strip('Node')
        if int(_node_no) > int(node_no) and item == 1:
            G.add_edge(node_no, _node_no)

nx.draw_networkx(G, node_color='k', edge_color='k', font_color='w')
plt.show()
```
